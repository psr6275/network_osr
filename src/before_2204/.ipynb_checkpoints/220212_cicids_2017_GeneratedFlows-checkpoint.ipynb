```python
import torch
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
filenames = glob.glob("../data/CICIDS2017/MachineLearningCVE/*")
filenames
```

```python
for i,fn in enumerate(filenames):
    print("load file",fn)
    if i==0:
        df = pd.read_csv(fn)
    else:
        df_temp = pd.read_csv(fn)
        df = pd.concat([df,df_temp])
        del df_temp
```

```python
df.head()
```

```python
print("the number of duplicates: ",len(df)-len(df.drop_duplicates()))
print("the ratio of duplicates: ",(len(df)-len(df.drop_duplicates()))/len(df))
```
