```python
labels = df[' Label'].unique()
lb_count = []
for lb in labels:
    lb_count.append(sum(df[' Label']==lb))
    print(sum(df[' Label']==lb))
```

```python
ax = df[' Label'].value_counts().plot(kind='bar')
for i in range(len(labels)):
    ax.text(i-0.2,100000, str(lb_count[i]), color='red', fontweight='bold')
```
